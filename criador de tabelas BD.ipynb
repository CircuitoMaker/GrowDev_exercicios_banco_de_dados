```python
import psycopg2

# função com os SQLs para criar astabelas no banco
def criar_tabelas(cur):
    # SQL para criar as tabelas
    comandos = (
        """
        CREATE TABLE IF NOT EXISTS Clientes (
            ID_Cliente SERIAL PRIMARY KEY,
            Nome VARCHAR(100),
            Endereco TEXT,
            Telefone VARCHAR(20),
            Email VARCHAR(100),
            Data_Nascimento DATE
        )
        """,
        """
        CREATE TABLE IF NOT EXISTS Produtos (
            ID_Produto SERIAL PRIMARY KEY,
            Nome VARCHAR(100),
            Descricao TEXT,
            Preco DECIMAL(10, 2),
            Quantidade_Estoque INT
        )
        """,
        """
        CREATE TABLE IF NOT EXISTS Pedidos (
            ID_Pedido SERIAL PRIMARY KEY,
            Data_Pedido TIMESTAMP,
            Status VARCHAR(20),
            Valor_Total DECIMAL(10, 2),
            ID_Cliente INT,
            FOREIGN KEY (ID_Cliente) REFERENCES Clientes(ID_Cliente)
        )
        """,
        """
        CREATE TABLE IF NOT EXISTS Itens_Pedido (
            ID_Item SERIAL PRIMARY KEY,
            Quantidade INT,
            Preco_Unitario DECIMAL(10, 2),
            Subtotal DECIMAL(10, 2),
            ID_Pedido INT,
            ID_Produto INT,
            FOREIGN KEY (ID_Pedido) REFERENCES Pedidos(ID_Pedido),
            FOREIGN KEY (ID_Produto) REFERENCES Produtos(ID_Produto)
        )
        """,
        """
        CREATE TABLE IF NOT EXISTS Pagamentos (
            ID_Pagamento SERIAL PRIMARY KEY,
            Data_Pagamento TIMESTAMP,
            Valor_Pago DECIMAL(10, 2),
            Metodo_Pagamento VARCHAR(50),
            ID_Pedido INT,
            FOREIGN KEY (ID_Pedido) REFERENCES Pedidos(ID_Pedido)
        )
        """
    )

    # chama a funçaõ que esecuta os comandos
    for comando in comandos:
        cur.execute(comando)

try:
    # conecta ao banco de dados
    conn = psycopg2.connect(
        dbname="ecommerce",
        user="postgres",
        password="postgres2",
        host="localhost",
        port="5432"
    )
    cur = conn.cursor()
    #verifica se a conexao com o bd aconteceu
    print("Conectado ao banco de dados!")

    # cria as tabelas
    criar_tabelas(cur)
    print("Tabelas criadas.... OK")

    # envia as alterações
    conn.commit()

except Exception as e:
    print(f"Falha ao conectar ao banco de dados ou criar tabelas: {e}")
finally:
    if cur:
        cur.close()
    if conn:
        conn.close()
```
