```python
import psycopg2

try:
    # Conectar ao banco de dados
    conn = psycopg2.connect(
        dbname="ecommerce",
        user="postgres",
        password="postgres2",
        host="localhost",
        port="5432"
    )
    cur = conn.cursor()

    # Consulta e exibição dos dados da tabela Clientes
    print("=== Tabela Clientes ===")
    cur.execute("SELECT * FROM Clientes LIMIT 10")
    rows = cur.fetchall()
    for row in rows:
        print(row)

    # Consulta e exibição dos dados da tabela Pedidos
    print("\n=== Tabela Pedidos ===")
    cur.execute("SELECT * FROM Pedidos LIMIT 10")
    rows = cur.fetchall()
    for row in rows:
        print(row)

    # Consulta e exibição dos dados da tabela Itens_Pedido
    print("\n=== Tabela Itens_Pedido ===")
    cur.execute("SELECT * FROM Itens_Pedido LIMIT 10")
    rows = cur.fetchall()
    for row in rows:
        print(row)

    # Consulta e exibição dos dados da tabela Produtos
    print("\n=== Tabela Produtos ===")
    cur.execute("SELECT * FROM Produtos LIMIT 10")
    rows = cur.fetchall()
    for row in rows:
        print(row)

    # Consulta e exibição dos dados da tabela Pagamentos
    print("\n=== Tabela Pagamentos ===")
    cur.execute("SELECT * FROM Pagamentos LIMIT 10")
    rows = cur.fetchall()
    for row in rows:
        print(row)

except Exception as e:
    print(f"Falha ao conectar ao banco de dados ou consultar dados: {e}")

finally:
    # Fechar o cursor e a conexão
    cur.close()
    conn.close()
    print("\nConexão encerrada.")
```
